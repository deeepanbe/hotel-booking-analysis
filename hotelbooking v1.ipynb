```python
import matplotlib.pyplot as plt

# Get correct counts and percentages
counts = df['is_canceled'].value_counts()
cancelled_perc = df['is_canceled'].value_counts(normalize=True)

# Sanity check (optional)
print(counts)

# Create plot
plt.figure(figsize=(5, 4))
plt.title('Reservation Status Count')

# Plot the actual counts
bars = plt.bar(['Not Canceled', 'Canceled'], counts, edgecolor='k', width=0.7)
plt.ylabel('Count')

# Add labels on top of bars
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2,
             height + (0.01 * height),   # small offset from top
             f'{counts[i]} ({cancelled_perc[i]*100:.1f}%)',
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()

```

```python

df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], dayfirst=True)
df['month'] = df['reservation_status_date'].dt.strftime('%B')


month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']


adr_by_hotel = df[df['is_canceled'] == 1].groupby(['hotel', 'month'])[['adr']].sum().reset_index()
adr_by_hotel['month'] = pd.Categorical(adr_by_hotel['month'], categories=month_order, ordered=True)
adr_by_hotel = adr_by_hotel.sort_values('month')


plt.figure(figsize=(14, 7))
ax = sns.barplot(data=adr_by_hotel, x='month', y='adr', hue='hotel', palette='Set2')
plt.title('ADR per Month for Canceled Bookings by Hotel Type', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Total ADR')
plt.xticks(rotation=45)
plt.legend(title='Hotel Type')

# Add value labels on top of bars
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{height:.0f}', 
                    (p.get_x() + p.get_width() / 2., height), 
                    ha='center', va='bottom', fontsize=9, color='black', xytext=(0, 5),
                    textcoords='offset points')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the data
df = pd.read_csv('hotel_bookings 2.csv')

# Step 2: Clean column names
df.columns = df.columns.str.strip().str.lower()

# Step 3: Filter canceled bookings
cancelled_data = df[df['is_canceled'] == 1]

# Step 4: Get top 10 countries by canceled bookings
top_10_country = cancelled_data['country'].value_counts().head(10)

# Step 5: Plot pie chart
plt.figure(figsize=(10, 8))
plt.pie(top_10_country, labels=top_10_country.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Countries with Most Canceled Reservations')
plt.axis('equal')  # Keeps the pie chart circular
plt.show()
```

```python
df.columns = df.columns.str.strip().str.lower()

# Step 3: Filter canceled bookings
cancelled_data = df[df['is_canceled'] == 1]

# Step 4: Count cancellations per market segment
market_segment_counts = cancelled_data['market_segment'].value_counts()

# Step 5: Plot pie chart
plt.figure(figsize=(10, 8))
plt.pie(market_segment_counts, labels=market_segment_counts.index, autopct='%2.3f%%', startangle=140)
plt.title('Canceled Bookings by Market Segment')
plt.axis('equal')  # Makes the pie chart a circle
plt.show()
```

```python

# Step 2: Clean column names
df.columns = df.columns.str.strip().str.lower()

# Step 3: Filter canceled bookings
cancelled_data = df[df['is_canceled'] == 1]

# Step 4: Calculate average ADR by market segment
adr_by_segment = cancelled_data.groupby('market_segment')['adr'].mean().sort_values(ascending=False)

# Step 5: Plot bar chart
plt.figure(figsize=(10, 6))
adr_by_segment.plot(kind='bar', color='skyblue')
plt.title('Average Daily Rate by Market Segment (Canceled Bookings)')
plt.ylabel('Average Daily Rate (ADR)')
plt.xlabel('Market Segment')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
df.columns = df.columns.str.strip().str.lower()

# Step 3: Create 'month' → 'quarter' mapping
month_to_quarter = {
    'January': 'Q1', 'February': 'Q1', 'March': 'Q1',
    'April': 'Q2', 'May': 'Q2', 'June': 'Q2',
    'July': 'Q3', 'August': 'Q3', 'September': 'Q3',
    'October': 'Q4', 'November': 'Q4', 'December': 'Q4'
}
df['quarter'] = df['arrival_date_month'].map(month_to_quarter)

# Step 4: Filter for canceled bookings
cancelled_data = df[df['is_canceled'] == 1]

# Step 5: Group by quarter and market segment, and calculate average ADR
adr_quarter_segment = cancelled_data.groupby(['quarter', 'market_segment'])['adr'].mean().reset_index()

# Step 6: Pivot for plotting
pivot_table = adr_quarter_segment.pivot(index='quarter', columns='market_segment', values='adr')

# Step 7: Plot
pivot_table.plot(kind='bar', figsize=(12, 6))
plt.title('Quarterly Average Daily Rate by Market Segment (Canceled Bookings)')
plt.ylabel('Average Daily Rate (ADR)')
plt.xlabel('Quarter')
plt.xticks(rotation=0)
plt.legend(title='Market Segment')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
